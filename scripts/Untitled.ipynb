```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.2 # true distribution samples
b = 0.6
true_dist = (a,b)

max_samples = 100
samples = np.random.uniform(true_dist[0], true_dist[1], (max_samples,))
```

```python
kl = []

for k in range(2,max_samples): # run through the size of the 'team' or sample size
    upper_bound_estimate = np.max(samples[:k]) # MLE estimate for upper bound
    lower_bound_estimate = np.min(samples[:k]) # MLE estimate for lower bound
    theta_estimate = upper_bound_estimate - lower_bound_estimate # Normalize range to make lower bound 0
    kl_div = np.log(theta_estimate / (b-a)) # estimate KL divergence between estimated distribution and true distribution
    kl.append(kl_div) # add kl divergence to list
    
plt.plot(kl) # plot kl divergence against how many samples were used
```
